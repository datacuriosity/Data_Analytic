```python
str_1 = "SELECT c.CUST_ID, concat(I.LAST_NAME, ' ', I.FIRST_NAME) AS FULL_NAME, c.ADDRESS FROM CUSTOMER c\
    INNER JOIN INDIVIDUAL I on c.CUST_ID = I.CUST_ID"
cust_1 = pd.read_sql(str_1, connection)

str_2 = "SELECT a.ACCOUNT_ID, a.CUST_ID, p.NAME AS PRODUCT_NAME, (a.AVAIL_BALANCE + a.PENDING_BALANCE) AS TOTAL_BALANCE \
FROM ACCOUNT \
a INNER JOIN PRODUCT P on a.PRODUCT_CD = P.PRODUCT_CD"
acc = pd.read_sql(str_2, connection)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Matplotlib

```python
x = np.linspace(-10, 10, 100)
plt.plot(x, np.sin(x), label='Sin func')
plt.plot(x, np.cos(x), label='Cosin func')
# plt.legend(loc='upper left')
plt.legend(loc=2)
plt.xlabel('Vales')
plt.ylabel('Function')
plt.title('Sin cosin plot')
plt.grid()
plt.show()
```

```python
x = np.linspace(-10, 10, 100)
y = np.random.rand(100)

plt.scatter(x, y)
plt.show()
```

```python
from seaborn import load_dataset
```

```python
tips = load_dataset('tips')
tips
```

```python
tips['size'].value_counts()
```

```python
plt.scatter(tips.total_bill, tips.tip, c='r', s=20, marker='*')
plt.scatter(tips['size'], tips.tip, c='b', s=40, marker='o', alpha=0.2)
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.show()
```

```python
plt.hist(tips.total_bill, label='total bill')
plt.hist(tips.tip, alpha=0.5, label='tip')
plt.legend(loc='best');
```

```python
ax_1 = plt.axes()
ax_1.hist(tips.total_bill, label='total bill')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(tips.total_bill, label='total bill')
plt.subplot(1, 2, 2)
plt.hist(tips.tip, alpha=0.5, label='tip')
plt.savefig('/Users/nguyen_truong/Desktop/his.png')
plt.show()
```

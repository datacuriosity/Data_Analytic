### Step 5. Create a pie chart presenting the male/female proportion

```python
males = (titanic['Sex'] == 'male').sum() / titanic.shape[0] * 100
males
females = (titanic['Sex'] == 'female').sum() / titanic.shape[0] * 100
females
proportion = [males, females]
plt.pie(x=proportion, colors=['blue', 'red'], shadow=False, explode = (0.15 , 0), startangle = 90, autopct = '%1.1f%%')
plt.title('Sex proportion')
plt.show()
```

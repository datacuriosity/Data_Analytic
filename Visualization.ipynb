```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Matplotlib

```python
x = np.linspace(-10,10,100)# tao mot array co cac khoang cach bang nhau
x
```

```python
x = np.linspace(-10,10,100)
plt.plot(x, np.sin(x), label='sin func')
plt.plot(x, np.cos(x), label='cos func')
plt.legend(loc='upper right')#bieu dien ten do thi
plt.xlabel('Values')
plt.ylabel('Function')
plt.title('Relationship sin and cos')
plt.grid()
plt.show()
```

## Scatter plot

```python
x = np.linspace(-10, 10, 100)
y = np.random.rand(100)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Moi quan he giua x va y')
plt.scatter(x, y)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips.total_bill, tips.tip, c='r', s=20, marker='v')
plt.scatter(tips['size'], tips.tip, c='b', s=40, marker='o', alpha=0.2)#alpha dung de xem cac diem du lieu o phia duoi khi cac diem du lieu chong len nhau
plt.xlabel('total_bill')
plt.ylabel('tip')
plt.show()
```

```python
tips['size'].value_counts()
```

```python
plt.hist(tips.total_bill);#hist la bieu do cot
plt.show()
plt.hist(tips.total_bill, label='total_bill')
plt.hist(tips.tip, alpha=0.2, label='tip')
plt.legend(loc='best');
```

```python
ax_1 = plt.axes()
ax_1.hist(tips.total_bill, label='total_bill');
```

```python
plt.figure(figsize=(10,6))#10 la chieu rong, 6 la chieu cao
plt.subplot(1, 2, 1)
plt.hist(tips.total_bill, label='total_bill')
plt.subplot(1, 2, 2)
plt.hist(tips.tip, alpha=0.2, label='tip');
plt.savefig('/Users/DELL/Documents/GitHub/Data_Analytic/hoang.png')#cach luu anh trong folder
plt.show()
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
sns.countplot(data=iris, x='species');
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length');
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', palette='dark');#hue la ma mau, to mau nhung diem nao thuoc ve loai hoa
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=iris, x='sepal_length', y='petal_length', hue='species', palette='dark', s=40);
```

```python
sns.displot(data=iris, x='petal_length', y='petal_width', height=9, aspect=0.8);#aspect la chieu rong
```

```python
sns.displot(data=iris, x='petal_length', kde=True, height=9, aspect=0.8);#kde ve duong uoc tinh di qua
```

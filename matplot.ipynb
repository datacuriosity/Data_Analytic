```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x))#mac dinh la ve ra scatter plot
```

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), label = 'sin')
plt.plot(x, np.cos(x), label = 'cos')
plt.legend(loc = 'upper right');
```

```python
x = np.linspace(0, 10, 1000)
plt.plot(x, np.sin(x), marker = 'v', label = 'sin')#marker la hinh dang do thi
plt.plot(x, np.cos(x), label = 'cos')
plt.legend(loc = 2)
plt.title('sin cos plot')
plt.xlabel('X label')
plt.ylabel('Y label');
```

```python
x = np.linspace(0, 10, 1000)
plt.figure(figsize=(10,6))#tang kich thuoc chieu x len 10, chieu y len 6
plt.plot(x, np.sin(x), c='r', label = 'sin')
plt.plot(x, np.cos(x), label = 'cos')
plt.legend(loc = 'upper right');
plt.title('sin cos plot')
plt.xlabel('X label')
plt.ylabel('Y label')
plt.savefig('sin_cos.png')#luu file anh
```

```python
plt.subplot(2, 1, 1)# tra ve axes dau tien
plt.plot(x, np.sin(x))
plt.title('Sin function')#ve 2 dong 1 cot
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
plt.title('Cos function');
```

```python
plt.figure(figsize=(15,6))
ax_1 = plt.subplot(1, 2, 1)# tra ve axes dau tien
ax_1.plot(x, np.sin(x))
ax_1.set_title('Sin function')#ve 1 dong 2 cot
ax_2 = plt.subplot(1, 2, 2)
ax_2.plot(x, np.cos(x))
ax_2.set_title('Cos function');
```

```python
fig, ax = plt.subplots(1, 2)
fig.set_figheight(6)
fig.set_figwidth(15)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

# Discrete: Countable(roi rac)
# Continuos: nhiet do, do luong, ap suat, can nang[0, 1]

```python
g = sns.FacetGrid(data= tips, row = 'sex', col = 'smoker')
```

```python
g = sns.FacetGrid(data= tips, row = 'sex', col = 'smoker', hue='time', margin_titles=True)
g.map(sns.scatterplot, 'total_bill', 'tip')
g.add_legend()
```

```python
g = sns.FacetGrid(data= tips, row = 'sex', col = 'smoker', hue='time', margin_titles=True)
g.map(sns.histplot, 'total_bill', 'tip')
g.add_legend()
```
